Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
options(repr.plot.width=4, repr.plot.height=5)

p = ggplot(penguins, aes(x = body_mass_g)) + 
    geom_histogram(na.rm = TRUE, binwidth = 100,color = "dodgerblue3", fill = "dodgerblue3", alpha = 0.75) + 
    theme_minimal() +
    facet_wrap(~year, ncol = 1) +
    scale_x_continuous(labels = comma_format()) + 
    labs(x = "Body Mass in Grams", y = "Number of Penguins", title = "Histogram of penguin body mass taken from 2007-2009") +
    theme(strip.text = element_text(hjust = 0.95, color = 'grey30', face = 'bold'))
p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.width=5, repr.plot.height=6, repr.plot.res = 300)

p = ggplot(penguins, aes(x = body_mass_g, y = sex)) +
    theme_minimal() +
    geom_point(na.rm = TRUE, alpha = 0.25, position = position_jitter(height = 0.2, width = 0)) +
    scale_x_continuous(labels = comma_format()) + 
    facet_wrap(~species, ncol = 1) +
    labs(y = '', x = 'Body Mass in Grams') +
    theme(strip.text = element_text(hjust = 0.05, color = 'grey30', face = 'bold'))
p
```
